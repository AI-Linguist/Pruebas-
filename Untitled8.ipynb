```python
import sqlite3
import xml.etree.ElementTree as ET
import re
from collections import Counter

def parse_segment_xml(xml_str):
    if not xml_str:
        return ""
    try:
        root = ET.fromstring(xml_str)
        parts = [v.text for v in root.iter()
                 if v.tag.endswith("Value") and v.text]
        return " ".join(parts).strip()
    except:
        values = re.findall(r"<Value>(.*?)</Value>", xml_str)
        return " ".join(values).strip()

def get_ngrams(tokens, n):
    return [" ".join(tokens[i:i+n]) for i in range(len(tokens)-n+1)]

def find_fragment_text(cursor, fhash, n_samples=50):
    """Find the actual phrase text behind a fragment hash."""
    cursor.execute("""
        SELECT tu.source_segment, tu.target_segment
        FROM translation_unit_fragments tuf
        JOIN translation_units tu ON tuf.translation_unit_id = tu.id
        WHERE tuf.fragment_hash = ?
        LIMIT ?
    """, (fhash, n_samples))
    rows = cursor.fetchall()
    if not rows:
        return None, None

    src_texts = [parse_segment_xml(r[0]) for r in rows]
    tgt_texts = [parse_segment_xml(r[1]) for r in rows]

    # Count n-grams across all samples
    # Fragment = n-gram present in at least 90% of samples
    threshold = int(len(src_texts) * 0.9)

    src_counter = Counter()
    tgt_counter = Counter()

    for src in src_texts:
        tokens = re.findall(r"\b[\w][\w\-]*\b", src.lower())
        seen = set()
        for n in range(2, 10):
            for ng in get_ngrams(tokens, n):
                if ng not in seen:
                    src_counter[ng] += 1
                    seen.add(ng)

    for tgt in tgt_texts:
        tokens = re.findall(r"\b[\w][\w\-]*\b", tgt.lower())
        seen = set()
        for n in range(2, 10):
            for ng in get_ngrams(tokens, n):
                if ng not in seen:
                    tgt_counter[ng] += 1
                    seen.add(ng)

    # Get candidates above threshold, longest first
    src_candidates = sorted(
        [(ng, cnt) for ng, cnt in src_counter.items() if cnt >= threshold],
        key=lambda x: (-len(x[0].split()), -x[1])
    )
    tgt_candidates = sorted(
        [(ng, cnt) for ng, cnt in tgt_counter.items() if cnt >= threshold],
        key=lambda x: (-len(x[0].split()), -x[1])
    )

    # Filter stop words from start/end
    EN_STOP = {"a","an","the","and","or","of","in","to","for","with","by",
               "at","on","from","as","is","are","was","were","be","it","its",
               "this","that","which","not","no","such","each","any","all"}
    ES_STOP = {"el","la","los","las","un","una","de","del","en","a","con",
               "por","para","que","y","o","se","su","sus","es","son","fue"}

    def is_content(ng, stops):
        t = ng.split()
        return t and t[0] not in stops and t[-1] not in stops

    src_best = [(ng, cnt) for ng, cnt in src_candidates
                if is_content(ng, EN_STOP)][:3]
    tgt_best = [(ng, cnt) for ng, cnt in tgt_candidates
                if is_content(ng, ES_STOP)][:3]

    return src_best, tgt_best

# Run on top 20 hashes
conn = sqlite3.connect("/content/Medical_en-GB_es-ES.sdltm")
cursor = conn.cursor()

top_hashes = [
    (2032267, 24123), (113010266601, 16695), (118115921, 16592),
    (108583977860843, 16536), (3504416510342, 15631), (3504730459525, 13506),
    (3507996004547, 12456), (3511303931839, 12294), (2032422, 11205),
    (113435748669, 10306), (3511286718841, 9618), (3500581140928, 9426),
    (3517200961585, 9117), (3514334415548, 9040), (3366686976296814, 8965),
    (113410594680, 8671), (3516929867112, 7991), (3644761264, 7703),
    (3366103906528769, 7453), (3500894132397, 7428)
]

print(f"{'HASH':<20} {'FREQ':>6}  {'EN FRAGMENT':<35} {'ES FRAGMENT':<35}")
print("-"*100)

for fhash, freq in top_hashes:
    src_best, tgt_best = find_fragment_text(cursor, fhash, n_samples=50)
    en = src_best[0][0] if src_best else "— not found —"
    es = tgt_best[0][0] if tgt_best else "— not found —"
    print(f"{fhash:<20} {freq:>6}  {en:<35} {es:<35}")

conn.close()
```

```python
import re

print("Reading file...")
with open("/content/sample_data/Medical_en-GB_es-ES.tmx", "rb") as f:
    content = f.read().decode("utf-8", errors="replace")

print(f"Total characters: {len(content):,}")

# Extract all valid TU blocks directly using regex
# This bypasses the XML parser entirely
print("Extracting TU blocks...")
tu_blocks = re.findall(r"<tu\b[^>]*>.*?</tu>", content, re.DOTALL)
print(f"Found {len(tu_blocks):,} TU blocks")

# Rebuild a clean TMX from scratch
header = '''<?xml version="1.0" encoding="UTF-8"?>
<tmx version="1.4">
  <header creationtool="RocheFix" datatype="plaintext"
          segtype="sentence" adminlang="en-GB" srclang="en-GB"/>
  <body>
'''
footer = "  </body>\n</tmx>"

print("Writing fixed TMX...")
with open("/content/sample_data/Medical_fixed.tmx", "w", encoding="utf-8") as f:
    f.write(header)
    for tu in tu_blocks:
        f.write("    " + tu.strip() + "\n")
    f.write(footer)

# Verify
import xml.etree.ElementTree as ET
try:
    tree = ET.parse("/content/sample_data/Medical_fixed.tmx")
    root = tree.getroot()
    count = sum(1 for _ in root.iter("tu"))
    print(f"✅ Valid XML — {count:,} translation units")
except ET.ParseError as e:
    print(f"❌ Still broken at: {e}")
```
